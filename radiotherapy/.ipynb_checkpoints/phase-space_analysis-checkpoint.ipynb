```python
data = psf.arrays()
E = data[b'Ekine']
print('Number of particles: ', len(E))
```

```python
n, bins, patches = plt.hist(E, 100, facecolor='green', alpha=0.75)
plt.title('Energy spectrum')
print('Energy min {} max {} mean {} std {}'.format(np.amin(E), np.amax(E), np.mean(E), np.std(E)))
```
